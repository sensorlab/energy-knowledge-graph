```python
import pandas as pd
import pickle
```

```python
df = pd.read_csv("../../data/raw/DRED/All_data.csv", skiprows=1, parse_dates=["Unnamed: 0"]).drop(columns=["unknown"])
df = df.rename(columns={'Unnamed: 0':'time'})
df = df.dropna()
```

```python
df["time"] = df["time"].astype(str)
df["time"] = df["time"].str.split("+").str[0]
```

```python
df["time"] = pd.to_datetime(df["time"])
df.reset_index(drop=True, inplace=True)
df.set_index("time", inplace=True)
```

```python
df.rename(columns={'mains':'aggregate'}, inplace=True)
```

```python
def watts2kWh(df):
    df = df /1000 * 1/3600
    return df
```

```python
df = watts2kWh(df)
```

```python
df
```

```python
data = {}
for c in df.columns:
    data[c] = pd.DataFrame(df[c])


res = {
    "DRED_1": data
}

# save to pickle
with open("DRED_processed.pickle", "wb") as f:
    pickle.dump(res, f, pickle.HIGHEST_PROTOCOL)

```
