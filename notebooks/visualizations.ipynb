# Exploratory data analysis

```python

import pandas as pd
import matplotlib.pyplot as plt
import os
import re
import numpy as np
# use latex font
plt.rc('text', usetex=True)
plt.rc('font', family='serif', size=15)

```

```python
def preprocess_string(string : str) -> str:
    string = string.lower().strip()
    string = re.sub(' +', ' ', string)
    string = string.replace("_", " ")
    string = string.replace("-", " ")
    string = string.replace("&", " ")
    string = string.split("(")[0]
    string = string.split("#")[0]

    string = string.strip()

    # handle known synoynms
    synonyms = {
        "refrigerator": "fridge",
        "vaccumcleaner": "vacuum cleaner",
        "breadmaker": "bread maker",
      
        
    }
    if "freezer" in string:
        string = "fridge"

    if string in synonyms:
        string = synonyms[string]

    if 'hi fi' in string:
        string = "audio system"

    if "router" in string:
        string = "router"

    if "treadmill" in string:
        string = "running machine"
        

    if "laptop" in string:
        string = "laptop"
    
    if "server" in string:
        string = "server"

    if "monitor" in string and not "baby" in string:
        string = "monitor"
    # special cases
    if "computer" in string and "charger" not in string:
        string = "pc"

    if "tv" in string:
        string = "television"

    if "television" in string:
        string = "television"

    if "macbook" in string:
        string = "laptop"
        
    if "car charger" == string:
        string = "ev"
    
    if "toast" in string:
        string = "toaster"
    
    if "modem" in string:
        string = "router"

    # we treat all audio devices as speakers so subwoofer is also a speaker
    if "subwoofer" in string:
        string = "speaker"

    if "speaker" in string:
        string = "speaker"

    if "iron" in string and "soldering" not in string:
        string = "iron"

    
    if "coffeemachine" in string:
        string = "coffee machine"
    if "coffee maker" in string:
        string = "coffee machine"

    if "dishwasher" in string:
        string = "dish washer"
    if "air conditioner" in string:
        string = "ac"

    if "air conditioning" in string:
        string = "ac"
    
    string = re.sub(' +', ' ', string)
    string = re.sub(r'\d+', '', string)
    return string.strip()
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import numpy as np
```

```python


df = pd.read_parquet("Energy_graph/energy-knowledge-graph/data/residential_metadata.parquet")
# get only unique coordinates
df = df.drop_duplicates(subset=["lat", "lon"])
```

```python
longitudes = df["lon"].values
latitudes = df["lat"].values
labels = df["country"].values

# longitudes = []
# latitudes = []
# labels = []

# manualy add the missing coordinates
capitals = {
    # "Switzerland": (46.8182, 8.2275),
    "Poland": (52.2297, 21.0122),
    "Netherlands": (52.1326, 5.2913),
    "Greece": (37.9838, 23.7275),
    "South Korea": (35.9078, 127.7669),
    "Germany": (51.1657, 10.4515)
}

for c in capitals:
    # if c not in labels:
        labels = np.append(labels, c)
        latitudes = np.append(latitudes, capitals[c][0])
        longitudes = np.append(longitudes, capitals[c][1])
```

```python


def plot_coordinates(longitudes, latitudes, labels=None):
    # Create a figure and a map
    plt.figure(figsize=(13, 7))
    m = Basemap(projection='merc', llcrnrlat=-57, urcrnrlat=75, llcrnrlon=-180, urcrnrlon=180, lat_ts=20, resolution='c')
    m.drawcoastlines(linewidth=0.5)
    m.drawcountries()
    m.fillcontinents(color='lightgray', lake_color='lightblue')
    m.drawmapboundary(fill_color='aqua')
    
    # Convert longitude and latitude to map projection coordinates
    x, y = m(longitudes, latitudes)
    
    # Plot coordinates
    m.scatter(x, y, marker='v', s=30, color="red")  # Use 's' to adjust marker size as needed
    
    if labels is not None:
        for label, xpt, ypt in zip(labels, x, y):
            plt.text(xpt, ypt, label, fontsize=12)
    # plt.title("World Map with Plotted Dataset Locations")
    plt.tight_layout()
    plt.savefig("plots/energy_graph/world_map.pdf", format="pdf")
    plt.show()

#write a sql query to get the coordinates of the cities


# # Example usage
# longitudes = [-74.0060, 2.3522, 151.2093]  # Example: New York, Paris, Sydney
# latitudes = [40.7128, 48.8566, -33.8688]
# labels = ['New York', 'Paris', 'Sydney']
plot_coordinates(longitudes, latitudes)
```
