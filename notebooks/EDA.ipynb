```python

house = "REFIT_1"
appliance = " television site"
def plot_histogram(data, house, appliance):
    df = data[house][appliance]
    # plot histogram for which hours the appliance is on during the day

    df = df.resample('H').mean()

    df = df.fillna(0)

    df = df.applymap(lambda x: 1 if x > 20 else 0)

    df = df.groupby(df.index.hour).sum()

    df.plot(kind='bar', figsize=(12, 6))
    plt.title(f'Time of Day when {appliance} is On in {house}')
    plt.xlabel('Hour of the Day')
    plt.xticks(rotation=0)
    plt.ylabel('Number of Days')

    plt.show()

plot_histogram(data, house, appliance)
plot_histogram(data, "REFIT_1", ' computer site')
plot_histogram(data, "REFIT_1",' chest freezer')
```

```python
import pandas as pd
import os



def convert_to_fraction_of_hour(sampling_rates):
    # Define conversion factors to seconds
    conversion_factors = {
        's': 1,
        'min': 60,
        'h': 3600
    }
    
    fractions = {}
    for key, value in sampling_rates.items():
        # Extract the number and the unit
        num = int(''.join(filter(str.isdigit, value)))
        unit = ''.join(filter(str.isalpha, value))
        print(key,num, unit, conversion_factors[unit])
        

        fractions[key] = (num * conversion_factors[unit])/3600
        
        
        
    return fractions


sampling_rates = {
    "UKDALE" : "6s",
    "ECO" : "1s",
    "REFIT" : "8s",
    "DRED" : "1s",
    "SUST2" : "2s",
    "HEART" : "1s",
    "DEKN" : "1min",
    "DEDDIAG" : "1s",
    "IDEAL" : "7s",
    "UCIML" : "1min",
    "LERTA" : "6s",
    "SUST1" : "1min",
    # US
    "REDD" : "1s",
    "HES" : "7s",
    "EEUD" : "1min",
    "HUE" : "1h",
    "ECDUY" : "15min",
    # ASIA
    "IAWE" : "1s",
    "ENERTALK" : "1s",
    "PRECON" : "1min",

}


sampling_rates = convert_to_fraction_of_hour(sampling_rates)

# watts to kWh given data frequency as a fraction of an hour (e.g. 0.5 for half-hourly data)
def watts2kwh(df: pd.Series, data_frequency: float) -> pd.Series:
    df = df / 1000 * data_frequency
    return df
sampling_rates
```

```python
import os 
from tqdm import tqdm
import pandas as pd

sampling_rates = {
    "UKDALE" : "6s",
    "ECO" : "1s",
    "REFIT" : "8s",
    "DRED" : "1s",
    "SUST2" : "2s",
    "HEART" : "1s",
    "DEKN" : "1min",
    "DEDDIAG" : "1s",
    "IDEAL" : "7s",
    "UCIML" : "1min",
    "LERTA" : "6s",
    "SUST1" : "1min",
    # US
    "REDD" : "1s",
    "HES" : "7s",
    "EEUD" : "1min",
    "HUE" : "1h",
    "ECDUY" : "15min",
    # ASIA
    "IAWE" : "1s",
    "ENERTALK" : "1s",
    "PRECON" : "1min",

}
import gc
data_availability_sampling = {}
# plot historgram of data availability for different sampling rates
for d in tqdm(os.listdir("./Energy_graph/energy-knowledge-graph/data/parsed")):
    print(d)
    if d.endswith(".pkl"):
        data = pd.read_pickle(f"./Energy_graph/energy-knowledge-graph/data/parsed/{d}")
        dataset_name = d.split(".")[0]
        curr_sampling_rate = sampling_rates[dataset_name]
        num_samples = 0
        for h in data:
            df = data[h]["aggregate"]
            num_samples += len(df)
        
        if curr_sampling_rate in data_availability_sampling:
            data_availability_sampling[curr_sampling_rate] += num_samples
        else:
            data_availability_sampling[curr_sampling_rate] = num_samples
        del data
        gc.collect()
            
```

```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sampling_rate_list= ["1s","2s", "6s", "7s", "8s", "1min", "15min", "1h"]

num_samples = [data_availability_sampling[s] for s in sampling_rate_list]

# plot cum sum histogram of data availability for different sampling rates

# Calculate the cumulative sum of the samples
cumulative_sum = np.cumsum(num_samples)



# Normalize to get percentages
total_samples = cumulative_sum[-1]  # Total number of samples
cumulative_percentage = (cumulative_sum / total_samples) * 100

# Plotting the cumulative percentage histogram
plt.figure(figsize=(10, 6))
sns.barplot(x=sampling_rate_list, y=cumulative_percentage, palette="Blues_d")
font_size = 20
# Add text annotations inside each bar
for i, percentage in enumerate(cumulative_percentage):
    if i == 0:
        plt.text(i, percentage-0.5, f'{percentage:.2f}%', ha='center', va='top', color='black', fontweight='bold', fontsize=font_size-7.9)
    elif i == 1:
        plt.text(i, percentage - 1, f'{percentage:.2f}%', ha='center', va='top', color='black', fontweight='bold', fontsize=font_size-7.9)
    elif i == 2:
        plt.text(i, percentage - 1, f'{percentage:.2f}%', ha='center', va='top', color='black', fontweight='bold', fontsize=font_size-7.9)
    else:
        plt.text(i, percentage - 5, f'{percentage:.2f}%', ha='center', va='top', color='white', fontweight='bold', fontsize=font_size-7.9)


# plt.title('Cumulative Percentage of Aggregate Consumption Data Availability for Different Sampling Rates')
plt.xlabel('Sampling Rate', fontsize=font_size)
plt.ylabel('Cumulative Data Availability (%)', fontsize=font_size)


# set ticks fontsize
plt.xticks(fontsize=font_size)
plt.yticks(fontsize=font_size)
plt.ylim(0, 100)  # Set y-axis limit to 100%
plt.tight_layout()
plt.savefig("figures/sampling_rate_availability_aggregate.pdf", format="pdf")
plt.show()

# Print the sampling rate list and number of samples
print(sampling_rate_list, num_samples)

```

```python
import pandas as pd 

import pickle
import re
from pathlib import Path

import numpy as np
import pandas as pd


def preprocess_string(string: str) -> str:
    string = string.lower().strip()
    string = re.sub(' +', ' ', string)
    string = string.replace("_", " ")
    string = string.replace("-", " ")
    string = string.replace("&", " ")
    string = string.split("(")[0]
    string = string.split("#")[0]

    string = string.strip()

    # handle known synonyms
    synonyms = {
        "refrigerator": "fridge",
        "vaccumcleaner": "vacuum cleaner",
        "breadmaker": "bread maker",

    }
    if "freezer" in string:
        string = "fridge"

    if string in synonyms:
        string = synonyms[string]

    if 'hi fi' in string:
        string = "audio system"

    if "router" in string:
        string = "router"

    if "treadmill" in string:
        string = "running machine"

    if "laptop" in string:
        string = "laptop"

    if "server" in string:
        string = "server"

    if "monitor" in string and "baby" not in string:
        string = "monitor"
    # special cases
    if "computer" in string and "charger" not in string:
        string = "pc"

    if "tv" in string:
        string = "television"

    if "television" in string:
        string = "television"

    if "macbook" in string:
        string = "laptop"

    if "car charger" == string:
        string = "ev"

    if "toast" in string:
        string = "toaster"

    if "modem" in string:
        string = "router"

    # we treat all audio devices as speakers so subwoofer is also a speaker
    if "subwoofer" in string:
        string = "speaker"

    if "speaker" in string:
        string = "speaker"

    if "iron" in string and "soldering" not in string:
        string = "iron"

    if "coffeemachine" in string:
        string = "coffee machine"
    if "coffee maker" in string:
        string = "coffee machine"

    if "dishwasher" in string:
        string = "dish washer"
    if "air conditioner" in string:
        string = "ac"

    if "air conditioning" in string:
        string = "ac"

    string = re.sub(' +', ' ', string)
    string = re.sub(r'\d+', '', string)
    return string.strip()


all_applicances = pd.read_pickle("Energy_graph/energy-knowledge-graph/data/temp/all_appliances.pkl")


all_applicances = list(all_applicances)
all_applicances.sort()
print("Number of appliances: ",len(all_applicances))
all_applicances


uniform_appliances = []

for a in all_applicances:
    uniform_appliances.append(preprocess_string(a))

uniform_appliances = set(uniform_appliances)
uniform_appliances = list(uniform_appliances)
uniform_appliances.sort()
print("Number of appliances: ",len(uniform_appliances))
uniform_appliances
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
def convert_to_fraction_of_hour(sampling_rates):
    # Define conversion factors to seconds
    conversion_factors = {
        's': 1,
        'min': 60,
        'h': 3600
    }
    
    fractions = {}
    for key, value in sampling_rates.items():
        # Extract the number and the unit
        num = int(''.join(filter(str.isdigit, value)))
        unit = ''.join(filter(str.isalpha, value))
        print(key,num, unit, conversion_factors[unit])
        

        fractions[key] = (num * conversion_factors[unit])/3600
        
        
        
    return fractions


sampling_rates = {
    "UKDALE" : "6s",
    "ECO" : "1s",
    "REFIT" : "8s",
    "DRED" : "1s",
    "SUST2" : "2s",
    "HEART" : "1s",
    "DEKN" : "1min",
    "DEDDIAG" : "1s",
    "IDEAL" : "7s",
    "UCIML" : "1min",
    "LERTA" : "6s",
    "SUST1" : "1min",
    # US
    "REDD" : "1s",
    "HES" : "7s",
    "EEUD" : "1min",
    "HUE" : "1h",
    "ECDUY" : "15min",
    # ASIA
    "IAWE" : "1s",
    "ENERTALK" : "1s",
    "PRECON" : "1min",

}


sampling_rates = convert_to_fraction_of_hour(sampling_rates)

# watts to kWh given data frequency as a fraction of an hour (e.g. 0.5 for half-hourly data)
def watts2kwh(df: pd.Series, data_frequency: float) -> pd.Series:
    df = df / 1000 * data_frequency
    return df
```

```python
%pip install tqdm
import numpy as np
import pandas as pd
from tqdm.notebook import tqdm
import os
import gc
PATH_TO_DATA = "./Energy_graph/energy-knowledge-graph/data/parsed/"






def count_values_in_bins(data_array, bins, existing_counts=None):
    """
    Counts the number of elements from data_array that fall into each bin defined by bins,
    and adds them to existing_counts if provided.

    Parameters:
    data_array (numpy.ndarray): The array of data to be counted into bins.
    bins (numpy.ndarray): The array defining the bin edges.
    existing_counts (numpy.ndarray, optional): An existing array of counts to add to.

    Returns:
    numpy.ndarray: An array of counts, where each count corresponds to the number of elements in a bin.
    """
    # Use numpy's digitize function to find the index of the bin for each element in data_array
    bin_counts, _ = np.histogram(data_array, bins=bins)

    # Count how many values fall into each bin
    # bin_counts = np.array([(bin_indices == i).sum() for i in range(1, len(bins))])

    # If existing counts are provided, add the new counts to them
    if existing_counts is not None:
        bin_counts = bin_counts + existing_counts

    return bin_counts


# 20kw
bins = np.linspace(0, 40000, 21)
# 2000w
# bins = np.linspace(0, 2000, 21)

#

bin_counts = None
sampling_rates = {
    "ECDUY" : 0.25,
    "HUE" : 1,
    "DEKN" : 1/60,
}
for d in tqdm(os.listdir(PATH_TO_DATA)):
    if d.endswith(".pkl"):
        # if d == "ECDUY.pkl":
        #     print("skipping", d)
        #     continue

        print(d)
        data = pd.read_pickle(f"{PATH_TO_DATA}/{d}")
        print("READ:", d)   
        if d == "ECDUY.pkl":
            print("processign ECDUY", d)
            for h in tqdm(data):
                df = data[h]["aggregate"]

                if "ECDUY.pkl" == d or "HUE.pkl" ==  d or "DEKN.pkl" == d:
                    # print(d, "converting to W")
                    curr_dataset= d.split(".")[0]
                    curr_sampling_rate = sampling_rates[curr_dataset]
                    df = df * 1000 / curr_sampling_rate
                    
                df.dropna(inplace=True)

                # # remove values lower than 5w
                # df = df[df > 5]

                
                # add counts to bins
                if bin_counts is None:
                    bin_counts = count_values_in_bins(df.values, bins)
                else:
                    bin_counts = count_values_in_bins(df.values, bins, bin_counts)
        else:
            for h in data:
                df = data[h]["aggregate"]

                if "ECDUY.pkl" == d or "HUE.pkl" ==  d or "DEKN.pkl" == d:
                    # print(d, "converting to W")
                    curr_dataset= d.split(".")[0]
                    curr_sampling_rate = sampling_rates[curr_dataset]
                    df = df * 1000 / curr_sampling_rate
                    
                df.dropna(inplace=True)

                # # remove values lower than 5w
                # df = df[df > 5]

                
                # add counts to bins
                if bin_counts is None:
                    bin_counts = count_values_in_bins(df.values, bins)
                else:
                    bin_counts = count_values_in_bins(df.values, bins, bin_counts)
        del data
        gc.collect()
```

```python
%pip install tqdm
import os
import gc
import numpy as np
import pandas as pd
from tqdm import tqdm
from numba import njit, prange
from concurrent.futures import ProcessPoolExecutor, as_completed

bins = np.linspace(0, 10, 20000)

bin_counts = None
sampling_rates = {
    "ECDUY" : 0.25,
    "HUE" : 1,
    "DEKN" : 1/60,
}
PATH_TO_DATA = "./Energy_graph/energy-knowledge-graph/data/parsed/"




def count_values_in_bins(data_array, bins):
    """
    Counts the number of elements from data_array that fall into each bin defined by bins,
    and adds them to existing_counts if provided.

    Parameters:
    data_array (numpy.ndarray): The array of data to be counted into bins.
    bins (numpy.ndarray): The array defining the bin edges.
    existing_counts (numpy.ndarray, optional): An existing array of counts to add to.

    Returns:
    numpy.ndarray: An array of counts, where each count corresponds to the number of elements in a bin.
    """
    # Use numpy's digitize function to find the index of the bin for each element in data_array
    bin_indices = np.digitize(data_array, bins)

    # Count how many values fall into each bin
    bin_counts = np.array([(bin_indices == i).sum() for i in range(1, len(bins)+1)])



    return bin_counts


def process_house(df, bins, sampling_rate):
    bin_counts = np.zeros(len(bins), dtype=np.int64)
    df_values = df.values
    df_values = df_values * 1000 / sampling_rate
    bin_counts = count_values_in_bins(df_values, bins)
    return bin_counts

def process_file(file_path, bins, sampling_rates):
    bin_counts = np.zeros(len(bins), dtype=np.int64)
    data = pd.read_pickle(file_path)
    dataset_name = os.path.basename(file_path).split(".")[0]
    
    # if dataset is ECDUY parallelize the processing in batches(one process runs 1/25th of the houses) data is a nested dictionary
    num_batches = 25
    if dataset_name == "ECDUY":
        return bin_counts
    else:
        print("Processing:", dataset_name)
        for h in data:
            df = data[h]["aggregate"]
            df_values = df.values.astype(np.float64)
            
            if dataset_name in ["ECDUY", "HUE", "DEKN"]:
                curr_sampling_rate = sampling_rates[dataset_name]
                df_values = df_values * 1000 / curr_sampling_rate
            
            # Compute bin counts
            bin_counts += count_values_in_bins(df_values, bins)
        
    del data
    gc.collect()
    return bin_counts

all_files = [os.path.join(PATH_TO_DATA, f) for f in os.listdir(PATH_TO_DATA) if f.endswith(".pkl")]

total_bin_counts = np.zeros(len(bins), dtype=np.int64)

# for file_path in tqdm(all_files):
#     print(file_path)
#     total_bin_counts += process_file(file_path, bins, sampling_rates)



    
print("Final bin counts:", total_bin_counts)

```

```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt

merged_lps = pd.read_pickle("Energy_graph/energy-knowledge-graph/data/loadprofiles/merged_loadprofiles.pkl")

all_aggregate_lps_daily = []
all_aggregate_lps_weekly = []
all_aggregate_lps_monthly = []

for h in tqdm(merged_lps):
    # if "ECDUY" in h:
    #     continue
    if merged_lps[h]["aggregate"]["daily"][0]> 10:
        continue


    if len(merged_lps[h]["aggregate"]["daily"].flatten()) != 24:
        continue

    if len(merged_lps[h]["aggregate"]["weekly"].flatten()) != 7:
        continue

    if len(merged_lps[h]["aggregate"]["monthly"].flatten()) != 31:
        continue
    all_aggregate_lps_daily.append(merged_lps[h]["aggregate"]["daily"].flatten())
    all_aggregate_lps_weekly.append(merged_lps[h]["aggregate"]["weekly"].flatten())
    all_aggregate_lps_monthly.append(merged_lps[h]["aggregate"]["monthly"].flatten())

# get the mean of all aggregate load profiles
mean_daily = np.mean(all_aggregate_lps_daily, axis=0)
mean_weekly = np.mean(all_aggregate_lps_weekly, axis=0)
mean_monthly = np.mean(all_aggregate_lps_monthly, axis=0)
```
