```python
# Home<letter>-meter<#>_<year>.csv


# HomeA_df_2014.drop(columns=['gen [kW]', "CellarOutlets [kW]", 'BedroomOutlets [kW]','MasterOutlets [kW]'])
test_str = "HomeF-meter2_2014.csv"
def parse_name(file: str):
    str = file.split("-")
    house = str[0]
    str = str[1].split("_")
    meter = str[0]
    year = str[1].split(".")[0]
    data = {
        "house": house,
        "meter": meter,
        "year": year
    }
    return data

data= parse_name(test_str)

print("house: "+ data["house"] + "\nmeter: " + data["meter"] + "\nyear: " + data["year"]) 
```

```python

data_dict = {}

homes = os.listdir("./Energy_graph/data/raw/SMART_HOME/")

for home in os.listdir("./Energy_graph/data/raw/SMART_HOME/"):
    if "Home" in home:
        home_data = {}
        for year in os.listdir("././Energy_graph/data/raw/SMART_HOME/"+home):
            for meter in os.listdir("././Energy_graph/data/raw/SMART_HOME/"+home+"/"+year):
                if "meter" in meter and meter.endswith(".csv"):
                    meter_parsed = parse_name(meter)
                    meter_data = pd.read_csv("././Energy_graph/data/raw/SMART_HOME/"+home+"/"+year+"/"+meter)
                    
                    useless_columns = [c for c in meter_data.columns if 'Outlets' in c or 'Receptables' in c]

                    useless_columns.append("gen [kW]")
                    meter_data.drop(columns=useless_columns, inplace=True)

                    home_data.setdefault(meter_parsed["meter"], []).append(meter_data)
    data_dict[home] = home_data
                     
                    

    break
        
```
