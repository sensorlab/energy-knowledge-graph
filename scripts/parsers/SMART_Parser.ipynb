```python
import pandas as pd
import os
```

```python
# convert to kWh
def convert2kWh(df):
    df['power[kW]'] = df['power[kW]'] * 0.25 # 15 min interval -> *4
    df.columns = ['time', 'power[kWh]']
    return df

# calcualte average consumption for given interval
def period_average_consumption(interval, df):
    df = df.resample(interval).sum()
    return df.mean()

# read the file and convert to kWh 
def process_file(file_path):
    df = pd.read_csv(file_path, header=None, names=["time", "power[kW]"])
    df = convert2kWh(df)
    df["time"] = pd.to_datetime(df["time"])
    df = df.set_index('time')
    return df

# get apartment name from file path
def apartment_name(file_path : str):
    return file_path.split("_")[0]

```

```python

folder_path = "../../data/SMART/"

# sets to store apartment names for each year
apartments_2014 = set()
apartments_2015 = set()
apartments_2016 = set()
 
for folder in os.listdir(folder_path):
    # iterate over all files in folder
    for file in os.listdir(folder_path + folder):
        if not file.endswith(".csv"):
            continue
        # read apartment name from file name
        apt_name = apartment_name(file)
        
        if folder == "2014":
            apartments_2014.add(apt_name)
        elif folder == "2015":
            apartments_2015.add(apt_name)
        elif folder == "2016":
            apartments_2016.add(apt_name)

        
# get the intersection of all three sets to get the apartments that appear in all three years
apartments = apartments_2014.intersection(apartments_2015, apartments_2016)


# sort apartments by number
apartments = sorted(list(apartments), key=sort_key)
apartments
```

```python
aptartments_2014_keys = set(apartments_2014.keys())
aptartments_2015_keys = set(apartments_2015.keys())
aptartments_2016_keys = set(apartments_2016.keys())


# get apartments that are in all three years
intersection = aptartments_2014_keys.intersection(aptartments_2015_keys)
intersection = intersection.intersection(aptartments_2016_keys)

```

```python
def sort_key(s):
    return int(s[3:])  # Extract the number part of the string and convert to int

# sort apartments by their number
apartments_list = sorted(list(intersection), key=sort_key)

# create df using apartments number as index
apt_consumption_df = pd.DataFrame(columns=["apartment_number","daily[kWh]", "weekly[kWh]", "monthly[kWh]"], index=apartments_list)
```

```python

curr_daily_avg = 0
curr_weekly_avg = 0
curr_monthly_avg = 0


# get average consumption for each apartment for all three years
for apt in apartments_list:
  
    curr_daily_avg = ((apartments_2014[apt][0] + apartments_2015[apt][0] + apartments_2016[apt][0]) / 3).values[0]
    curr_weekly_avg = ((apartments_2014[apt][1] + apartments_2015[apt][1] + apartments_2016[apt][1]) / 3).values[0]
    curr_monthly_avg = ((apartments_2014[apt][2] + apartments_2015[apt][2] + apartments_2016[apt][2]) / 3).values[0]

    
    # put data in dataframe
    apt_consumption_df.loc[apt] = [apt,curr_daily_avg, curr_weekly_avg, curr_monthly_avg]
  
apt_consumption_df.head()
```
