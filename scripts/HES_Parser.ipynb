```python
# process data for one day and return it as a dataframe
def process_file(file_name):
    # get date from file name
    date = os.path.basename(file_name).split(".")[0]
    
    with open(file_name, "r") as f:
        # read file 
        HES = pd.read_csv(f, header=None, sep=";")

        # name columns
        HES.columns = ["time", "sensor", "consumption"]
        
        # add date to time column
        HES["time"] = date + " " + HES["time"]

        # time column to datetime
        HES["time"] = pd.to_datetime(HES["time"], format="%Y-%m-%d %H:%M:%S")
        HES = process_table(HES)

    return HES


# pivots the table so that each appliance has its own column with consumption values and adds a column for total consumption
def process_table(df):
    # replace sensor number with appliance name
    df["sensor"] = df["sensor"].replace(appliance_dict)

    # pivot the table so that each appliance has its own column with consumption values
    df = df.pivot(index='time', columns='sensor', values='consumption')

    # add column for total consumption
    df["Total Consumption"] = df.sum(axis=1)  

    df = df.rename_axis(None, axis=1)
    return df

    
process_file("../data/HES/July/2018-07-01.csv").head()
```
